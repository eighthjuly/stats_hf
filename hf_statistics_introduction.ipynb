**A tale of two charts**

```python
import pandas as pd
import matplotlib.pyplot as plt
month = ['Jul', 'Aug' ,'Sep', 'Oct', 'Nov', 'Dec']
profit = [2.0, 2.1, 2.2, 2.1, 2.3, 2.4]
plt.style.use('ggplot')
plt.figure(figsize=(12,6))
plt.suptitle('Company Profit')
plt.subplot(1,2,1)
plt.plot(month, profit, label= 'profit')
plt.ylabel("Profit (in millions)", labelpad=10 )
plt.ylim(0, 2.5)
plt.legend()
plt.subplot(1,2,2)
plt.plot(month, profit, label= 'profit')
plt.ylabel("Profit (in millions)", labelpad=10 )
plt.legend()
plt.show()
```

Both charts are based on the same data,but they each send a different message. The first chart shows that the profit is steady. It achieves this by having the vertical axis start at 0, and then plotting the profit for each month against this. The second chart gives a different impression by making the vertical axis start at a different place and adjusting the scale accordingly. At a glance, the profits appear to be rising dramatically each month.

Software can translate data into charts, but it’s up to you to make 
sure the chart is right.

**The humble pie chart**

Pie charts divide your entire data set into distinct groups.

Example: Pie chart showing the number of units sold per genre

Pie charts show proportions

```python
genre = ['Sports','Strategy','Action','Shooter','Other']
unit_sold = [27500, 11500, 6000, 3500, 1500]
colors = ["#355070", "#6d597a", "#b56576", "#e56b6f", "#eaac8b"]
df = pd.DataFrame({'genre': genre, 'unit_sold': unit_sold })
plt.figure(figsize = (8,8))
plt.style.use('fivethirtyeight')
plt.title('Units sold per genre', fontsize=10, c="black") 
plt.pie(df['unit_sold'], labels= df['genre'], colors =colors, autopct ='%.0f%%', explode=(0,0,0,0,.1))
plt.show()
```

For data with proportions less vation, we go for bar charts. 
**Bar charts can allow for more accuracy**

Just like pie charts, bar charts allow you to compare relative sizes, but the advantage of using a bar chart is that they allow for a greater degree of precision. Bar charts can be drawn either vertically or horizontally.

**Vertical bar charts**

Vertical bar charts show categories on the horizontal axis.

An example showing the sales figures in units for five regions, A, B, C, D, and E:

```python
region = ['A','B','C','D','E']
sales = [1000,5000,7500,8000,9500]
plt.bar(region, sales)
plt.xlabel('Region')
plt.ylabel('Sales(units)')
plt.show()
```

**Horizontal bar charts**

Horizontal bar charts are just like vertical bar charts except that the axes are flipped round.

Vertical bar charts tend to be more common, but horizontal bar charts are useful if the names of your categories are long. They give you lots of space for showing the name of each category without having to turn the bar labels sideways.

```python
genre = ['Sports','Strategy','Action','Shooter','Other']
satisfied_in_percentage = [99,90,85,95,80]
percentage_label = ['20%','40%','60%','80%','100%']
plt.barh(genre,satisfied_in_percentage, color = 'olive')
plt.xticks([20,40,60,80,100], labels = percentage_label)
plt.show()
```

**Dealing with multiple sets of data**

With bar charts, it’s actually really easy to show more than one set of data on the same chart. In this case, we can use split category and segmented bar.

Below is the example of plants data when it germinated and died.


```python
plants = ['Spinach', 'Turnip', 'Rhubarb', 'Broccoli', 'Kale']
died = [10,25,5,30,21]
germinated = [74, 88, 56,69,59]
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.title("The split-category bar chart")
plt.barh(plants, died, color= 'red')
plt.barh(plants, germinated,color = 'green', height=0.4, align = 'edge')
plt.xlabel("Number Of Seeldings")
plt.tight_layout()
plt.subplot(1,2,2)
plt.title("The segmented bar chart")
plt.barh(plants, germinated, color='green' )
plt.barh(plants, died,color = 'red')
plt.xlabel("Number Of Seeldings")
plt.tight_layout()
plt.show()
```

**Categories vs. Numbers**

When you’re working with charts, one of the key things you need to figure out is what sort of data you’re dealing with. Once you’ve figured that out, you’ll find it easier to make key decisions about what chart you need to best represent your data.

**Categorical or qualitative data:** 
The data is split into categories that describe qualities or characteristics. For this reason, it’s also called qualitative
data. An example of qualitative data is game genre; each genre forms a separate category. The key thing to remember with qualitative data is that the data values can’t be interpreted as numbers.

**Numerical or quantitative data:** 
Numerical data, on the other hand, deals with numbers. It’s data where the values have meaning as numbers, and that involves measurements or counts. Numerical data is also called quantitative data because it describes quantities.
